```python
import numpy as np
import matplotlib.pyplot as plt
X=np.array([1,2,3,4,5,6,7])
Y=np.array([2,4,5,4,5,7,8])
X=X[:,np.newaxis]
Y=Y[:,np.newaxis]
print(X.shape,Y.shape)
```

```python
plt.scatter(X,Y)
```

```python
theta=np.random.randn(2,1)
```

```python
theta
```

```python
alpha=0.01
iterations=10000
```

```python
X=np.hstack((np.ones(X.shape),X))
X.shape
```

```python
theta.shape
```

```python
def h_theta(a,b):
    return a.dot(b)
```

```python
print(h_theta(X,theta))
```

```python
def cost_function(a,b,c):
    return 1/(2*len(a))*(np.sum(h_theta(a,c)-b)**2)
```

```python
print(cost_function(X,Y,theta))
```

```python
def gradient(a,b,c):
    return 1/len(a)*(a.T.dot(h_theta(a,c)-b))
```

```python
print(gradient(X,Y,theta))
```

```python
def gradient_descent(a,b,c,d,e):
    cost_history=np.zeros(e)
    new_theta=c
    for i in range(0,e):
        new_theta=new_theta-d*gradient(a,b,new_theta)
        cost_history[i]=cost_function(a,b,new_theta)
    return new_theta,cost_history
```

```python
new_theta,cost_history=gradient_descent(X,Y,theta,alpha,iterations)
print(cost_history)
print(new_theta)
```

```python
pred=h_theta(X,new_theta)
pred
```

```python
plt.scatter(X[:,1],Y,color="blue")
plt.plot(X[:,1],pred,color="red")
plt.show()
```
