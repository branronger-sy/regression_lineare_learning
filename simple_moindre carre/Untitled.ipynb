```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
X,Y=make_regression(n_samples=100,n_features=1,noise=10)
plt.scatter(X,Y)
```

```python
def moy(a):
    return 1/len(a)*np.sum(a)
```

```python
def var(a):
    return 1/len(a)*np.sum(a**2)-moy(a)
```

```python
def ecart(a):
    return var(a)**(1/2)
```

```python
def cov(a,b):
    if not len(a)==len(b):
        print("les parametre dovent etre de meme shape")
    return 1/len(a)*np.sum(a*b)-moy(a)*moy(b)
```

```python
def theta_finder(a,b):
    theta1=cov(a,b)/ecart(a)**2
    theta0=moy(b)-theta1*moy(a)
    return theta0,theta1
```

```python
def h_theta(a,theta0,theta1):
    return theta0+theta1*a
```

```python
a=np.linspace(min(X),max(X),100)
theta1,theta2=theta_finder(X,Y)
b=h_theta(a,theta1,theta2)
```

```python
plt.scatter(X,Y)
plt.plot(a,b)
```
