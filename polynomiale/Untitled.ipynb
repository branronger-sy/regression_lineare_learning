```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
x,y=make_regression(n_samples=100,n_features=1,noise=10)
print(x.shape)
y=y+abs(y/2)
Y=y[:,np.newaxis]
print(Y.shape)
plt.scatter(x,Y)
```

```python
theta=np.random.randn(5,1)
print(theta)
```

```python
X=np.hstack((np.ones(x.shape),x,x**2,x**3,x**4))
print(X.shape)
print(X)
```

```python
def h_theta(a,b):
    return a.dot(b)
```

```python
print(h_theta(X,theta))
plt.scatter(x,Y)
plt.plot(x,h_theta(X,theta),c='r')
```

```python
def cost_function(a,b,c):
    return 1/(2*len(a))*np.sum((h_theta(a,c)-b)**2)
```

```python
print(cost_function(X,Y,theta))
```

```python
def gradient(a,b,c):
    return 1/len(a)*a.T.dot(h_theta(a,c)-b)
```

```python
def gradient_descent(a,b,c,d,e):
    cost_history=np.zeros(e)
    new_theta=c.copy()
    for i in range(0,e):
        new_theta=new_theta-d*gradient(a,b,new_theta)
        cost_history[i]=cost_function(a,b,new_theta)
    return new_theta,cost_history
```

```python
new_theta,cost_history=gradient_descent(X,Y,theta,0.001,10000)
```

```python
print(new_theta)
```

```python
idx = np.argsort(x[:,0])

plt.scatter(x,Y)
plt.plot(x[idx], h_theta(X,new_theta)[idx], c='r')
```

```python
plt.figure()
plt.plot(cost_history)
plt.title("Cost Function")
plt.show()
```
